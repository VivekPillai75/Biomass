```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from osgeo import gdal
from PIL import Image as im
```

```python
df=pd.read_csv("features_metadata_FzP19JI.csv")
df.head()
```

```python
dataset=df[["filename","split"]]
```

```python
dataset.head()
```

```python
for i in range(len(dataset)):
    imageName=dataset["filename"].iloc[i]
    if(dataset["split"].iloc[i]=="test"):
        if(imageName.split("_")[1]=="S1"):
            fileLocation="test_features/"+imageName
            ds=gdal.Open(fileLocation)
            
            finalArray=np.zeros((256,256))
            for i in range(1,5):
                data_array= ds.GetRasterBand(i).ReadAsArray()
                npDataArray=np.array(data_array)
                normData = npDataArray/np.linalg.norm(npDataArray)
                finalArray=finalArray+normData
            finalArray=finalArray/4
        else:
            fileLocation="test_features/"+imageName
            ds=gdal.Open(fileLocation)
            
            finalArray=np.zeros((256,256))
            for i in range(1,12):
                data_array= ds.GetRasterBand(i).ReadAsArray()
                npDataArray=np.array(data_array)
                normData = npDataArray/np.linalg.norm(npDataArray)
                finalArray=finalArray+normData
            finalArray=finalArray/11
        data=im.fromarray(finalArray)
        file="testFeatures/"+imageName
        data.save(file)
```
